```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import folium

#read the csv file
data=pd.read_csv('D:\Dataset\kavi.csv')
print(data)

# Handle missing values (if any)
data.dropna(inplace=True)

# Convert data types if necessary
data['Temperature'] = pd.to_numeric(data['Temperature'])
data['Humidity'] = pd.to_numeric(data['Humidity'])
data['WindSpeed'] = pd.to_numeric(data['WindSpeed'])
data['GeneralDiffuseFlows'] = pd.to_numeric(data['GeneralDiffuseFlows'])
data['DiffuseFlows'] = pd.to_numeric(data['DiffuseFlows'])
data['PowerConsumption_Zone1'] = pd.to_numeric(data['PowerConsumption_Zone1'])
data['PowerConsumption_Zone2'] = pd.to_numeric(data['PowerConsumption_Zone2'])
data['PowerConsumption_Zone3'] = pd.to_numeric(data['PowerConsumption_Zone3'])


print(data.describe())

#SCATTER PLOT
plt.figure(figsize=(8, 6))
plt.scatter(data['Temperature'], data['PowerConsumption_Zone1'], alpha=0.5, label='Zone 1')
plt.scatter(data['Temperature'], data['PowerConsumption_Zone2'], alpha=0.5, label='Zone 2')
plt.scatter(data['Temperature'], data['PowerConsumption_Zone3'], alpha=0.5, label='Zone 3')
plt.title('Temperature vs. Power Consumption')
plt.xlabel('Temperature')
plt.ylabel('Power Consumption')
plt.legend()
plt.show()

# Plot energy usage over time for Power Consumption Zones
plt.figure(figsize=(12, 6))
for zone in ['PowerConsumption_Zone1', 'PowerConsumption_Zone2', 'PowerConsumption_Zone3']:
    plt.plot(data.index, data[zone], label=zone)

plt.title('Energy Usage Over Time for Power Consumption Zones')
plt.xlabel('Time')
plt.ylabel('Power Consumption')
plt.grid(True)
plt.legend()
plt.show()

# Plot environmental factors (Temperature, Humidity, WindSpeed) over time using an area chart
plt.figure(figsize=(12, 6))
plt.fill_between(data.index, data['Temperature'], color='yellow', alpha=0.3, label='Temperature')
plt.fill_between(data.index, data['Humidity'], color='teal', alpha=0.3, label='Humidity')
plt.fill_between(data.index, data['WindSpeed'], color='red', alpha=0.3, label='WindSpeed')

plt.title('Environmental Factors Over Time')
plt.xlabel('Time')
plt.ylabel('Value')
plt.grid(True)
plt.legend()
plt.show()


# Sample energy consumption data (replace with your actual dataset)
data = {
    'Building': ['Building A', 'Building B', 'Building C', 'Building D'],
    'PowerConsumption_Zone1': [1500, 2200, 1800, 2500],  # Energy consumption values for Zone 1
    'PowerConsumption_Zone2': [1200, 1900, 1500, 2200],  # Energy consumption values for Zone 2
    'PowerConsumption_Zone3': [1000, 1600, 1300, 1900]   # Energy consumption values for Zone 3
}

# Create a DataFrame from the sample data
df = pd.DataFrame(data)

# Display the DataFrame
print(df)


# Set the index to 'Building' column for plotting
df.set_index('Building', inplace=True)

# Plotting a bar chart for energy consumption across buildings (Zones)
df.plot(kind='bar', figsize=(10, 6))
plt.title('Energy Consumption Across Buildings (Zones)')
plt.xlabel('Building')
plt.ylabel('Energy Consumption (kWh)')
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.legend(title='Zone')
plt.grid(axis='y', linestyle='--', alpha=0.5)  # Add gridlines for y-axis
plt.show()

# Sample data (replace with your actual dataset)
data = {
    'Temperature': [25, 28, 30, 22, 24],
    'Humidity': [60, 55, 50, 65, 70],
    'WindSpeed': [10, 12, 8, 15, 7],
    'GeneralDiffuseFlows': [100, 120, 90, 110, 105],
    'DiffuseFlows': [80, 85, 75, 90, 88],
    'PowerConsumption_Zone1': [1500, 1400, 1600, 1350, 1450],
    'PowerConsumption_Zone2': [1200, 1300, 1100, 1250, 1150],
    'PowerConsumption_Zone3': [1000, 1050, 950, 1100, 1020]
}

# Create DataFrame from sample data
df = pd.DataFrame(data)

# Compute correlation matrix
corr_matrix = df.corr()

# Plot heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of Variables')
plt.show()  

# Sample data (replace with your actual dataset)
data = {
    'TimePeriod': ['Period1', 'Period2', 'Period3', 'Period4', 'Period5'],
    'PowerConsumption_Zone1': [1500, 1400, 1600, 1350, 1450],
    'PowerConsumption_Zone2': [1200, 1300, 1100, 1250, 1150],
    'PowerConsumption_Zone3': [1000, 1050, 950, 1100, 1020]
}

# Create DataFrame from sample data
df = pd.DataFrame(data)

# Set the 'TimePeriod' column as the index for plotting
df.set_index('TimePeriod', inplace=True)

# Plot stacked column chart
plt.figure(figsize=(10, 6))
df.plot(kind='bar', stacked=True, colormap='viridis', alpha=0.8)

# Adding labels and title
plt.title('Power Consumption Across Zones Over Time')
plt.xlabel('Time Period')
plt.ylabel('Power Consumption (kWh)')
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.legend(title='Zone')

# Display the stacked column chart
plt.show()

# Sample data (replace with your actual dataset)
data = {
    'Zone': ['Zone1', 'Zone2', 'Zone3'],
    'PowerConsumption': [1500, 1200, 1000]  # Energy consumption values for each zone
}

# Create DataFrame from sample data
df = pd.DataFrame(data)

# Plot pie chart
plt.figure(figsize=(8, 8))
plt.pie(df['PowerConsumption'], labels=df['Zone'], autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightgreen', 'orange'], explode=(0.1, 0, 0))
plt.title('Energy Consumption Across Zones')

# Display the pie chart
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.show()
```
